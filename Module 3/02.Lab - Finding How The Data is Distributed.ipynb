### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
print("Columns:")
print(df.columns)
print("\nData Types:")
print(df.dtypes)
print("\nSummary Information:")
print(df.info())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
print("\nMissing Values:")
print(df.isnull().sum())

df_cleaned = df.dropna()
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
print("\nEmployment Distribution:")
print(df['Employment'].value_counts())

print("\nJob Satisfaction Distribution:")
print(df['JobSat'].value_counts())

print("\nYears of Professional Coding:")
print(df['YearsCodePro'].value_counts())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
plt.figure(figsize=(10, 6))
df['JobSat'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Job Satisfaction Distribution')
plt.ylabel('')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
# First, check the data type and handle potential missing values
print(df['LanguageHaveWorkedWith'].dtype)
print(df['LanguageHaveWorkedWith'].isnull().sum())

# Function to safely split languages
def split_languages(languages):
    if pd.isna(languages):
        return []
    return str(languages).split(';')

# Split the languages into lists
df['LanguagesHave'] = df['LanguageHaveWorkedWith'].apply(split_languages)
df['LanguagesWant'] = df['LanguageWantToWorkWith'].apply(split_languages)

# Flatten the lists to count languages
have_langs = [lang for sublist in df['LanguagesHave'] for lang in sublist]
want_langs = [lang for sublist in df['LanguagesWant'] for lang in sublist]

# Create visualization
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
pd.Series(have_langs).value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Languages Worked With')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
pd.Series(want_langs).value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Languages Wanted')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Print top languages for verification
print("\nTop Languages Worked With:")
print(pd.Series(have_langs).value_counts().head())
print("\nTop Languages Wanted:")
print(pd.Series(want_langs).value_counts().head())
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
plt.figure(figsize=(10, 6))
df.groupby(['Country'])['RemoteWork'].value_counts(normalize=True).unstack().plot(kind='bar', stacked=True)
plt.title('Remote Work Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Proportion')
plt.legend(title='Remote Work', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
# Create a mapping for job satisfaction
job_sat_mapping = {
    'Very dissatisfied': 1, 
    'Dissatisfied': 2, 
    'Neutral': 3, 
    'Satisfied': 4, 
    'Very satisfied': 5
}

# Create a mapping for years of professional coding
years_coding_mapping = {
    'Less than 1 year': 0,
    '1 to 2 years': 1.5,
    '3 to 4 years': 3.5,
    '5 to 9 years': 7,
    '10 to 14 years': 12,
    '15 to 19 years': 17,
    '20 or more years': 25
}

# Convert JobSat and YearsCodePro to numeric
df['JobSatNumeric'] = df['JobSat'].map(job_sat_mapping)
df['YearsCodeProNumeric'] = df['YearsCodePro'].map(years_coding_mapping)

# Calculate correlation
correlation = df['JobSatNumeric'].corr(df['YearsCodeProNumeric'])
print("\nCorrelation between Job Satisfaction and Years of Professional Coding:")
print(correlation)
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
cross_tab = pd.crosstab(df['Employment'], df['EdLevel'], normalize='index')
plt.figure(figsize=(12, 6))
cross_tab.plot(kind='bar', stacked=True)
plt.title('Employment Status by Education Level')
plt.xlabel('Employment Status')
plt.ylabel('Proportion')
plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
