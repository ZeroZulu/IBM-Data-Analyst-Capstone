**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
## Write your code here
plt.figure(figsize=(10, 6))
plt.hist(pd.read_sql_query("SELECT CompTotal FROM main WHERE CompTotal IS NOT NULL", conn), 
         bins=20, edgecolor='black')
plt.title('Distribution of Total Compensation')
plt.xlabel('Total Compensation')
plt.ylabel('Frequency')
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
## Write your code here
query = """
SELECT CAST(Age AS NUMERIC) AS Age 
FROM main 
WHERE Age IS NOT NULL 
AND Age != '' 
AND Age != 'NA'
"""

# Read the query and ensure numeric type
age_data = pd.read_sql_query(query, conn)
age_data['Age'] = pd.to_numeric(age_data['Age'], errors='coerce')

# Remove any remaining non-numeric values
age_data = age_data.dropna()

plt.figure(figsize=(10, 6))
plt.boxplot(age_data['Age'])
plt.title('Distribution of Age')
plt.ylabel('Age')
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here
df_scatter = pd.read_sql_query("SELECT Age, WorkExp FROM main WHERE Age IS NOT NULL AND WorkExp IS NOT NULL", conn)

plt.figure(figsize=(10, 6))
plt.scatter(df_scatter['Age'], df_scatter['WorkExp'], alpha=0.5)
plt.title('Age vs Work Experience')
plt.xlabel('Age')
plt.ylabel('Work Experience')
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here
query = """
SELECT 
    CAST(TimeSearching AS NUMERIC) AS TimeSearching, 
    CAST(Frustration AS NUMERIC) AS Frustration, 
    CAST(Age AS NUMERIC) AS Age 
FROM main 
WHERE 
    TimeSearching IS NOT NULL AND TimeSearching != '' AND TimeSearching != 'NA'
    AND Frustration IS NOT NULL AND Frustration != '' AND Frustration != 'NA'
    AND Age IS NOT NULL AND Age != '' AND Age != 'NA'
"""

# Read the data
df_bubble = pd.read_sql_query(query, conn)

# Convert columns to numeric, coercing errors to NaN
df_bubble['TimeSearching'] = pd.to_numeric(df_bubble['TimeSearching'], errors='coerce')
df_bubble['Frustration'] = pd.to_numeric(df_bubble['Frustration'], errors='coerce')
df_bubble['Age'] = pd.to_numeric(df_bubble['Age'], errors='coerce')

# Drop any rows with NaN values
df_bubble = df_bubble.dropna()

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(df_bubble['TimeSearching'], df_bubble['Frustration'], 
            s=df_bubble['Age'], alpha=0.5)
plt.title('Time Searching vs Frustration (Bubble Size: Age)')
plt.xlabel('Time Searching')
plt.ylabel('Frustration')
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
df_pie = pd.read_sql_query("""
SELECT DatabaseWantToWorkWith, COUNT(*) as count 
FROM main 
WHERE DatabaseWantToWorkWith IS NOT NULL
GROUP BY DatabaseWantToWorkWith 
ORDER BY count DESC 
LIMIT 5
""", conn)

plt.figure(figsize=(10, 6))
plt.pie(df_pie['count'], labels=df_pie['DatabaseWantToWorkWith'], autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Want to Work With')
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
## Write your code here
query = """
SELECT Age, TimeSearching, TimeAnswering
FROM main 
WHERE Age BETWEEN 30 AND 35
"""

# Read the data
df = pd.read_sql_query(query, conn)

# Calculate median for each age
df_stacked = df.groupby('Age').agg({
    'TimeSearching': 'median',
    'TimeAnswering': 'median'
}).reset_index()

# Create the stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar(df_stacked['Age'], df_stacked['TimeSearching'], label='Time Searching')
plt.bar(df_stacked['Age'], df_stacked['TimeAnswering'], 
        bottom=df_stacked['TimeSearching'], label='Time Answering')
plt.title('Median Time Searching and Answering (Age 30-35)')
plt.xlabel('Age')
plt.ylabel('Time')
plt.legend()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
query = """
SELECT Age, CompTotal
FROM main 
WHERE Age BETWEEN 45 AND 60
"""

# Read the data
df = pd.read_sql_query(query, conn)

# Calculate median for each age
df_line = df.groupby('Age')['CompTotal'].median().reset_index()

# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(df_line['Age'], df_line['CompTotal'], marker='o')
plt.title('Median Total Compensation by Age (45-60)')
plt.xlabel('Age')
plt.ylabel('Median Total Compensation')
plt.grid(True)
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here
df_hbar = pd.read_sql_query("""
SELECT MainBranch, COUNT(*) as count 
FROM main 
WHERE MainBranch IS NOT NULL
GROUP BY MainBranch 
ORDER BY count DESC
""", conn)

plt.figure(figsize=(10, 6))
plt.barh(df_hbar['MainBranch'], df_hbar['count'])
plt.title('Distribution of Main Branch')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.tight_layout()
plt.show()
```
