### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
# Check the data types and handle potential issues
print(df['SOPartFreq'].dtype)
print(df['JobSat'].dtype)

# Convert columns to numeric, handling any potential non-numeric values
def safe_numeric_convert(series):
    return pd.to_numeric(series, errors='coerce')

# Clean the data
df['SOPartFreq_Numeric'] = safe_numeric_convert(df['SOPartFreq'])
df['JobSat_Numeric'] = safe_numeric_convert(df['JobSat'])
df['Age_Numeric'] = safe_numeric_convert(df['Age'])

# Remove rows with NaN values
df_cleaned = df.dropna(subset=['SOPartFreq_Numeric', 'JobSat_Numeric', 'Age_Numeric'])

# Create the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['Age_Numeric'], 
            df_cleaned['SOPartFreq_Numeric'], 
            s=df_cleaned['JobSat_Numeric']*20, 
            alpha=0.6)
plt.title('Age vs. Participation Frequency')
plt.xlabel('Age')
plt.ylabel('Participation Frequency')
plt.colorbar(label='Job Satisfaction')
plt.tight_layout()
plt.show()

# Print some diagnostic information
print("Original data shape:", df.shape)
print("Cleaned data shape:", df_cleaned.shape)
print("\nData types after conversion:")
print(df_cleaned[['Age_Numeric', 'SOPartFreq_Numeric', 'JobSat_Numeric']].dtypes)
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
# Comprehensive data cleaning and conversion function
def prepare_numeric_column(series):
    # Convert to string first to handle any unusual formatting
    series_str = series.astype(str)
    
    # Remove any non-numeric characters
    series_clean = series_str.str.replace(r'[^\d.]', '', regex=True)
    
    # Convert to numeric
    return pd.to_numeric(series_clean, errors='coerce')

# Clean the data
df['ConvertedCompYearly_Numeric'] = prepare_numeric_column(df['ConvertedCompYearly'])
df['JobSat_Numeric'] = prepare_numeric_column(df['JobSat'])
df['Age_Numeric'] = prepare_numeric_column(df['Age'])

# Remove rows with NaN or zero values
df_cleaned = df[
    (df['ConvertedCompYearly_Numeric'].notna()) & 
    (df['ConvertedCompYearly_Numeric'] > 0) &
    (df['JobSat_Numeric'].notna()) & 
    (df['Age_Numeric'].notna())
]

# Normalize bubble sizes
def normalize_bubble_size(series, min_size=20, max_size=500):
    # Ensure positive values
    series_positive = series - series.min() + 1
    return (series_positive / series_positive.max()) * (max_size - min_size) + min_size

# Create the bubble plot
plt.figure(figsize=(12, 7))
scatter = plt.scatter(
    df_cleaned['ConvertedCompYearly_Numeric'], 
    df_cleaned['JobSat_Numeric'], 
    s=normalize_bubble_size(df_cleaned['Age_Numeric']), 
    alpha=0.6,
    c=df_cleaned['Age_Numeric'],  # Color by age
    cmap='viridis'
)
plt.title('Compensation vs. Job Satisfaction')
plt.xlabel('Yearly Compensation')
plt.ylabel('Job Satisfaction')
plt.colorbar(scatter, label='Age')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Diagnostic information
print("Data Analysis:")
print("Total rows in original dataset:", len(df))
print("Rows after cleaning:", len(df_cleaned))
print("\nCompensation Summary:")
print(df_cleaned['ConvertedCompYearly_Numeric'].describe())
print("\nJob Satisfaction Summary:")
print(df_cleaned['JobSat_Numeric'].describe())
print("\nAge Summary:")
print(df_cleaned['Age_Numeric'].describe())
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
import numpy as np

# Comprehensive data cleaning function
def prepare_numeric_column(series):
    return pd.to_numeric(series.astype(str).str.replace(r'[^\d.]', '', regex=True), errors='coerce')

# Clean numeric columns
df['Age_Numeric'] = prepare_numeric_column(df['Age'])
df['JobSat_Numeric'] = prepare_numeric_column(df['JobSat'])

# Function to safely split and count languages
def count_languages(languages_str):
    if pd.isna(languages_str):
        return []
    return str(languages_str).split(';')

# Identify top programming languages
def get_top_languages(df, column, top_n=5):
    # Flatten all languages
    all_languages = [lang for sublist in df[column].apply(count_languages) for lang in sublist]
    # Count and get top languages
    return pd.Series(all_languages).value_counts().head(top_n).index.tolist()

# Get top 5 languages
top_languages = get_top_languages(df, 'LanguageHaveWorkedWith')

# Prepare plot
plt.figure(figsize=(12, 8))

# Color palette
colors = plt.cm.Set3(np.linspace(0, 1, len(top_languages)))

# Plot for each language
for i, language in enumerate(top_languages):
    # Filter for rows containing this language
    language_df = df[df['LanguageHaveWorkedWith'].apply(lambda x: language in str(x) if pd.notna(x) else False)]
    
    # Ensure we have clean numeric data
    language_df = language_df.dropna(subset=['Age_Numeric', 'JobSat_Numeric'])
    
    # Bubble size based on job satisfaction
    bubble_sizes = language_df['JobSat_Numeric'] * 20
    
    plt.scatter(
        language_df['Age_Numeric'], 
        language_df['JobSat_Numeric'], 
        s=bubble_sizes, 
        alpha=0.6, 
        label=language,
        color=colors[i]
    )

plt.title('Programming Language Preferences by Age and Job Satisfaction')
plt.xlabel('Age')
plt.ylabel('Job Satisfaction')
plt.legend(title='Languages', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
# Comprehensive data cleaning function
def prepare_numeric_column(series):
    return pd.to_numeric(series.astype(str).str.replace(r'[^\d.]', '', regex=True), errors='coerce')

# Clean numeric columns
df['ConvertedCompYearly_Numeric'] = prepare_numeric_column(df['ConvertedCompYearly'])
df['JobSat_Numeric'] = prepare_numeric_column(df['JobSat'])

# Function to safely split and count databases
def count_databases(databases_str):
    if pd.isna(databases_str):
        return []
    return str(databases_str).split(';')

# Identify top preferred databases
def get_top_databases(df, column, top_n=5):
    # Flatten all databases
    all_databases = [db for sublist in df[column].apply(count_databases) for db in sublist]
    # Count and get top databases
    return pd.Series(all_databases).value_counts().head(top_n).index.tolist()

# Get top 5 preferred databases
top_databases = get_top_databases(df, 'DatabaseWantToWorkWith')

# Prepare plot
plt.figure(figsize=(12, 8))

# Color palette
colors = plt.cm.Set2(np.linspace(0, 1, len(top_databases)))

# Plot for each database
for i, database in enumerate(top_databases):
    # Filter for rows containing this database
    database_df = df[df['DatabaseWantToWorkWith'].apply(lambda x: database in str(x) if pd.notna(x) else False)]
    
    # Ensure we have clean numeric data
    database_df = database_df.dropna(subset=['ConvertedCompYearly_Numeric', 'JobSat_Numeric'])
    
    # Bubble size based on count of database preference
    bubble_sizes = database_df['DatabaseWantToWorkWith'].apply(lambda x: str(x).count(database) * 50)
    
    plt.scatter(
        database_df['JobSat_Numeric'], 
        database_df['ConvertedCompYearly_Numeric'], 
        s=bubble_sizes, 
        alpha=0.6, 
        label=database,
        color=colors[i]
    )

plt.title('Preferred Databases vs. Job Satisfaction and Compensation')
plt.xlabel('Job Satisfaction')
plt.ylabel('Yearly Compensation')
plt.legend(title='Databases', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
# Comprehensive data cleaning function
def prepare_numeric_column(series):
    return pd.to_numeric(series.astype(str).str.replace(r'[^\d.]', '', regex=True), errors='coerce')

# Clean numeric columns
df['ConvertedCompYearly_Numeric'] = prepare_numeric_column(df['ConvertedCompYearly'])
df['JobSat_Numeric'] = prepare_numeric_column(df['JobSat'])

# Process developer roles
def process_dev_roles(roles_str):
    if pd.isna(roles_str):
        return []
    return [role.strip() for role in str(roles_str).split(';')]

# Clean and prepare data
df['DevRoles'] = df['DevType'].apply(process_dev_roles)

# Explode the roles to create separate rows
df_exploded = df.explode('DevRoles')

# Remove rows with invalid data
df_cleaned = df_exploded[
    (df_exploded['ConvertedCompYearly_Numeric'].notna()) & 
    (df_exploded['ConvertedCompYearly_Numeric'] > 0) &
    (df_exploded['JobSat_Numeric'].notna())
]

# Get top roles
top_roles = df_cleaned['DevRoles'].value_counts().head(5).index.tolist()

# Prepare plot
plt.figure(figsize=(12, 8))

# Color palette
colors = plt.cm.Set1(np.linspace(0, 1, len(top_roles)))

# Plot for each role
for i, role in enumerate(top_roles):
    # Filter for specific role
    role_df = df_cleaned[df_cleaned['DevRoles'] == role]
    
    # Calculate bubble sizes based on job satisfaction
    bubble_sizes = role_df['JobSat_Numeric'] * 30
    
    plt.scatter(
        role_df['ConvertedCompYearly_Numeric'], 
        role_df['JobSat_Numeric'], 
        s=bubble_sizes, 
        alpha=0.6, 
        label=role,
        color=colors[i]
    )

plt.title('Compensation vs. Job Satisfaction Across Developer Roles')
plt.xlabel('Yearly Compensation')
plt.ylabel('Job Satisfaction')
plt.legend(title='Developer Roles', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
# Comprehensive data cleaning function
def prepare_numeric_column(series):
    return pd.to_numeric(series.astype(str).str.replace(r'[^\d.]', '', regex=True), errors='coerce')

# Clean numeric columns
df['ConvertedCompYearly_Numeric'] = prepare_numeric_column(df['ConvertedCompYearly'])
df['JobSat_Numeric'] = prepare_numeric_column(df['JobSat'])

# Function to safely split and count web frameworks
def count_web_frameworks(frameworks_str):
    if pd.isna(frameworks_str):
        return []
    return str(frameworks_str).split(';')

# Identify top preferred web frameworks
def get_top_web_frameworks(df, column, top_n=5):
    # Flatten all web frameworks
    all_frameworks = [fw for sublist in df[column].apply(count_web_frameworks) for fw in sublist]
    # Count and get top web frameworks
    return pd.Series(all_frameworks).value_counts().head(top_n).index.tolist()

# Get top 5 preferred web frameworks
top_web_frameworks = get_top_web_frameworks(df, 'WebframeWantToWorkWith')

# Prepare plot
plt.figure(figsize=(12, 8))

# Color palette
colors = plt.cm.Set3(np.linspace(0, 1, len(top_web_frameworks)))

# Plot for each web framework
for i, framework in enumerate(top_web_frameworks):
    # Filter for rows containing this framework
    framework_df = df[df['WebframeWantToWorkWith'].apply(lambda x: framework in str(x) if pd.notna(x) else False)]
    
    # Ensure we have clean numeric data
    framework_df = framework_df.dropna(subset=['ConvertedCompYearly_Numeric', 'JobSat_Numeric'])
    
    # Bubble size based on framework preference frequency
    bubble_sizes = framework_df['WebframeWantToWorkWith'].apply(lambda x: str(x).count(framework) * 50)
    
    plt.scatter(
        framework_df['JobSat_Numeric'], 
        framework_df['ConvertedCompYearly_Numeric'], 
        s=bubble_sizes, 
        alpha=0.6, 
        label=framework,
        color=colors[i]
    )

plt.title('Preferred Web Frameworks vs. Job Satisfaction and Compensation')
plt.xlabel('Job Satisfaction')
plt.ylabel('Yearly Compensation')
plt.legend(title='Web Frameworks', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
def create_admired_technologies_bubble_plot(df):
    # Handle NA values in 'LanguageAdmired' column
    df['LanguageAdmired'] = df['LanguageAdmired'].fillna('')
    
    # Get top 5 countries
    top_countries = df['Country'].value_counts().head(5).index
    
    # Get top 5 admired languages
    # Use string split and stack to handle multiple languages
    top_languages_admired = df['LanguageAdmired'].str.split(';', expand=True).stack().value_counts().head(5).index
    
    # Create the plot
    plt.figure(figsize=(15, 8))
    
    # Track colors for each language
    color_map = plt.cm.get_cmap('tab10')
    
    for i, country in enumerate(top_countries):
        country_df = df[df['Country'] == country]
        
        for j, lang in enumerate(top_languages_admired):
            # Careful string matching to handle multiple languages
            subset = country_df[country_df['LanguageAdmired'].str.contains(lang, na=False)]
            
            if not subset.empty:
                # Count occurrences of the language
                lang_counts = subset['LanguageAdmired'].str.count(lang)
                
                plt.scatter(
                    [country]*len(subset), 
                    subset['JobSat'], 
                    s=lang_counts*20,  # Size based on language count
                    label=f'{country} - {lang}', 
                    alpha=0.6,
                    color=color_map(j/len(top_languages_admired))
                )
    
    plt.title('Admired Technologies Across Top Countries')
    plt.xlabel('Country')
    plt.ylabel('Job Satisfaction')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()


create_admired_technologies_bubble_plot(df)
```
